# read( ) &  readlines( ) & BOM & readline( )


```python
# read
f=open("sample_data/o.txt","r",encoding="UTF-8")
str1=f.read(5)
print(str1)
f.close()
```

```python
# readlines
with open("sample_data/o.txt","r",encoding="UTF-8") as f:
  content=f.readlines()
  print(type(content))
  print(content)
```

```python
# BOM
# UTF-8 ,Windows 系統中儲存為UTF-8編碼時,檔案的最前方自動產生占了1個字元的BOM(Unicode Byte Order Mark), 識別文字為UTF-8
with open("sample_data/o.txt","r",encoding="UTF-8") as f:
  content=f.readlines()
  print(type(content))
  print(content)

f=open("sample_data/o.txt","r",encoding="UTF-8")
str1=f.read(5)
print(str1)
f.close()  # BOM \ufeff
```

```python
with open("sample_data/o.txt","r",encoding="UTF-8-sig") as f:
  content=f.readlines()
  print(type(content))
  print(content)

f=open("sample_data/o.txt","r",encoding="UTF-8-sig")
str1=f.read(5)
print(str1)
f.close()  # BOM \ufeff
```

```python
# readline
f=open("sample_data/o.txt","r",encoding="UTF-8-sig")
print(f.readline())
print(f.readline())
print(f.readline())
f.close()  # BOM \ufeff
```

```python
# try except finally
try:
  a=int(input("number 1 :"))
  b=int(input("number 2 :"))
  r=a+b
  print("r=",r)
except:
  print("Input Number Error!")
```

```python
# try except finally
try:
  a=int(input("number 1 :"))
  b=int(input("number 2 :"))
  r=a%b
  print("r=",r)
except ValueError:
  print("Input Number Error!")
except Exception as e:
  print(e)
finally:
  print("finally code.")
```
