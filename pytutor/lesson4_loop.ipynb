```python
import time
for i in range(1,10,1):
  print("Loop %2d :" % i,end="")
  for j in range(1,(10-i)):
    print(" ",end="") # 加空格
  for k in range(1,i+1):
    print(k,"",end="") # 數字加空間
    time.sleep(0.3)
  print("") # 換行
```

```python
numA=1
numB=1
for i in range(1,10,1):
  for j in range(1,10,1):
    print("%2d x %2d = %2d " % (numA,numB,numA*numB)," ",end="")
    numB+=1
  numA+=1
  numB=1
  print("")
```
