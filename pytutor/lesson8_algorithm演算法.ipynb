# 找最大 最小值

```python
# 找最大 最小值
datas=[4,3,6,5,2]
minv=datas[0]
maxv=datas[0]
for item in datas:
  if minv>item:
    minv=item
  if maxv<item:
    maxv=item
  pass
print("min = %d, max = %d" % (minv,maxv) )
```

```python
# 找最大 最小值
datas=[4,3,6,5,2]
minidx=0
maxidx=0
for i in range(datas):
  if datas[minidx]>datas[i]:
    minidx=i
  if datas[maxidx]<datas[i]:
    maxidx=i
  pass
print("min = %d, max = %d" % (datas[minidx],datas[maxidx]) )
```

# 二分搜尋
![](https://res.cloudinary.com/practicaldev/image/fetch/s--7HtJ-as8--/c_imagga_scale,f_auto,fl_progressive,h_900,q_auto,w_1600/https://dev-to-uploads.s3.amazonaws.com/i/5hsod7t93v85b23rk671.png)

```python
def BubbleSort(d,r=None):
  n=len(d)-1
  for i in range(n):
    for j in range(n-i):
      if d[j]>d[j+1] :
        # python 可以使用 swap two values
        d[j] , d[j+1] = d[j+1] , d[j]
        # 關聯串列處理
        if not r==None: r[j] , r[j+1] = r[j+1] , r[j]

# numb -ref- name
numb=[256,   731, 943,  389, 142, 645,  829,  946]
name=["小虎","中森","木淼","大同","子孔","美麗","温柔","來多"]

BubbleSort(numb,name)
print(numb)
print(name)


nu=int(input("input number:"))
# BinarySearch
# Start
L=0      # index low Boundary
H=len(numb)-1 # index high Boundary
IsFound=-1   # IsFound
while(L<=H):
  Mid=(L+H)//2    # 注意地方
  print("before",L,H,Mid)
  if(numb[Mid]==nu):
    IsFound=Mid
    break
  elif(numb[Mid]>nu):
    H=Mid-1    # 注意地方
  else:
    L=Mid+1    # 注意地方
  print("after",L,H,Mid)
# End

if IsFound>-1:
  print(numb[IsFound],name[IsFound])
else:
  print("No Rec.")

```
