### key鍵是唯一
出現重覆時,會被覆蓋   
dict1={"chn":90,"eng":80,"mat":70,"chn":100}     
dict1["chn"] # 100    
鍵不存在時    
dict1["gc"]  # Error 程式會中斷     
使get     
dict1.get("gc") # None   
dict1.get("gc",-1) # -1       




```python
# A 內向 B 外向 O 自信 AB 聰明
# coding
blood_character={"A" :"內向", "B":"外向", "O": "自信", "AB": "聰明","A":"內向型"}
blood_character["A"] # ? 內向 / 內向型





```

```python
blood_character["OOO"] # Error
blood_character.get("OOO") # None
blood_character.get("OOO","未定義") # 未定義
```
