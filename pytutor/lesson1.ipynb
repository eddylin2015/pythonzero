## 一.前課重點
binary file 二進制文件, bin文件是機器碼.   
10100...   128位指令, 可能是表達 a + 1.       
所以高級(高階)程式表達方法,接近自然語言a = a+1。底階接近機器方式,   
MOV @a, REG1;    # 10100...   128位指令    
ADD REG1, 1;     # 10101...   128位指令    
MOV REG1; @a;    # 10100...   128位指令
