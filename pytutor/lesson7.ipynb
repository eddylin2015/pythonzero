# 變數有效範圍


```python
def scope():
  var1=1 #def var1
  print(var1,var2)  #1,20  hua ge(home)  hua ge(he kou hua)
var1=10
var2=20
scope()
print(var1,var2)    #10,20
```

```python
def scope():
  global var1    #hua ge(out)
  var1=1
  print(var1,var2)  #1,20
var1=10
var2=20
scope()
print(var1,var2)    #1,20
```
