```python
"""
Least common multiples(LCM)
"""
mx=0
a=int(input("input number 1:"))
b=int(input("input number 2:"))
if a<b :
  mx=b
else:
  mx=a
for i in range(mx,a*b+1,1):
  if i%a==0 and i%b==0:
    print("最小公倍數是",i,".")
    print("")
    break
```

```python
mx=1
"""
largest common factor(LCF)
"""
sl=1
a=int(input("input number 1  :"))
b=int(input("input number 2  :"))
if a<b :
  a=mx
else :
  b=mx
for i in range(mx,sl-1,-1):
  if a%i==0 and b%i==0:
    print("最大公因數是",i)
    break
```
